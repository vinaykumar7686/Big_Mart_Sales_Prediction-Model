## Segregating Dependent and Independent Variables

```python
#reading the cleaned version of data into cdf variable
cdf=pd.read_csv('train_bm_Cleaned.csv')
cdf.shape
```

```python
x=cdf.drop(['Item_Outlet_Sales'], axis=1)
```

```python
y=cdf["Item_Outlet_Sales"]
x.shape,y.shape
```

## Scaling of variables

```python
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()
scaledx = scaler.fit_transform(x) 
```

```python
x=pd.DataFrame(scaledx, columns= x.columns)
```

## Dividind dataframe into train-test dataframes

```python
from sklearn.model_selection import train_test_split
trainx, testx, trainy, testy = train_test_split(x,y, random_state= 56)
```

## Implementing KNN

```python
from sklearn.neighbors import KNeighborsRegressor as KNN
from sklearn.metrics import mean_squared_error as MSE
from sklearn.metrics import mean_absolute_error as MAE
```

```python
def KNNAlgo(n=5):
    reg=KNN(n_neighbors= n)

    #Training Model
    reg.fit(trainx, trainy)

    #Predicting
    res = reg.predict(testx)

    MSerr=MSE(res, testy)
    MAerr=MAE(res, testy)

    #print(f"Mean Squared Error: {MSerr}")
    #print(f"Mean Absolute Error: {MAerr}")
    return MSerr

KNNAlgo()
```

## Finding best value of k 

```python
errs=[]
for i in range(3,40,1):
    errs.append(KNNAlgo(i))
```

```python
plt.plot(range(3,40,1), errs)
plt.show()
```

```python
# From above Graph it is evident that this algorithworks at its best when value of k is about 7
KNNAlgo(n=7)
```
